### 寫一個DEF

```python
import matplotlib.pyplot as plt

def plot_cod_alk_relation(dates, COD, alk, title):
    fig, ax1 = plt.subplots(figsize=(10,6))

    # 繪製鹼度(直方圖)
    ax1.bar(dates, alk, color='#FF4500', alpha=0.5, label='鹼度(CaCO3 mg/L)')
    ax1.set_xlabel('日期', fontsize=18)
    ax1.set_ylabel('鹼度 (CaCO3 mg/L)', color='#FF4500', fontsize=18)
    ax1.tick_params('y', colors='#FF4500', labelsize=14)

    # 根據數據調整 y 軸範圍
    ax1.set_ylim(min(alk) - 10, max(alk) + 10)

    # 建立第二 y 軸
    ax2 = ax1.twinx()
    ax2.plot(dates, COD, color='g', marker='o', label='COD去除效率(%)')
    ax2.set_ylabel('COD去除效率(%)', color='g', fontsize=18)
    ax2.tick_params('y', colors='g', labelsize=14)
    ax2.set_ylim(70, 100)

    # 設定圖表標題和圖例
    plt.title(title, fontsize=18)
    lines, labels = ax1.get_legend_handles_labels()
    lines2, labels2 = ax2.get_legend_handles_labels()
    ax1.legend(lines + lines2, labels + labels2, loc='upper left')

    # 顯示圖表
    plt.tight_layout()
    plt.show()

# 第一組數據
dates_1 = ['11/8', '11/10']
COD_1 = [93.25 ,95.5]
alk_1 = [109.48, 131.53]
plot_cod_alk_relation(dates_1, COD_1, alk_1, '對照組鹼度與 COD 之關係')

# 第二組數據
dates_2 = ['11/14', '11/16', '11/21', '11/23']
COD_2 = [91.46, 96.26, 98.52, 96.81]
alk_2 = [327.66, 335.41, 336.88, 311.96]
plot_cod_alk_relation(dates_2, COD_2, alk_2, '第一組鹼度與 COD 之關係')
```
